```python
from urllib.parse import quote_plus

import pandas as pd
from sqlalchemy import create_engine

username = quote_plus("root")
password = quote_plus("password")
server = "host.docker.internal:3306"
db_name = "TESTDB"

conn = create_engine(
    f"mysql+pymysql://{username}:{password}@{server}/{db_name}",
).connect()
```

```python
df = pd.read_sql(
    "select * from Staff",
    conn,
)
df
```

```python
df.dtypes
```

```python
df["ID"] = df["ID"].apply(lambda x: x.replace("01", "31"))
df
```

```python
df.to_sql(
    "staff",
    conn,
    if_exists="append",
    index=False
)

df
```
